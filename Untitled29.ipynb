```python
# Rossmann - Full Pipeline (fixed XGBoost predict + learning curves + SHAP)
# Path: /content/drive/MyDrive/Rossmann Stores Data (2).csv
# This script is an enhanced, debugged version of your pipeline:
# - fixes XGBoost predict ntree_limit -> iteration_range
# - safer best_iteration handling
# - LightGBM num_iteration handling
# - plots learning curves for XGBoost and LightGBM
# - computes feature importance and SHAP explanations (if shap installed)
# - saves artifacts

import os, warnings, json
warnings.filterwarnings("ignore")

# safe mount helper (Colab)
try:
    from google.colab import drive
    if not os.path.exists("/content/drive/MyDrive"):
        drive.mount('/content/drive', force_remount=False)
    else:
        print("Drive appears mounted already.")
except Exception as e:
    print("Not running in Colab or mount failed (proceeding assuming drive already available).", e)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
import joblib

# LightGBM & helpers
import lightgbm as lgb
from lightgbm import early_stopping, log_evaluation

# xgboost (will use xgb.train)
import xgboost as xgb

# optional CatBoost import (safe)
try:
    from catboost import CatBoostRegressor
    CATBOOST_AVAILABLE = True
    print("CatBoost available.")
except Exception:
    CATBOOST_AVAILABLE = False
    print("CatBoost NOT available. Skipping CatBoost training (you can `!pip install catboost` to enable).")

# SHAP (optional) - try import, else set flag
try:
    import shap
    SHAP_AVAILABLE = True
except Exception:
    SHAP_AVAILABLE = False
    print("SHAP not available. To enable, run: pip install shap")

# -------------------------
# 1. Config & paths
# -------------------------
RANDOM_STATE = 42
DATA_PATH = "/content/drive/MyDrive/Rossmann%20Stores%20Data%20%282%29.csv"
ARTIFACT_DIR = "/content/drive/MyDrive/rossmann_artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)
DATE_SPLIT = "2015-06-01"   # chronological test split (adjustable)

# -------------------------
# 2. Load data
# -------------------------
print("Loading data:", DATA_PATH)
df = pd.read_csv(DATA_PATH, parse_dates=['Date'], low_memory=False)
print("Initial shape:", df.shape)
print("Columns:", df.columns.tolist())

# -------------------------
# 3. Basic cleaning & sanity
# -------------------------
df['Open'] = df.get('Open', pd.Series(1, index=df.index)).fillna(1).astype(int)
df = df[df['Open'] == 1].copy()   # keep open days
df['StateHoliday'] = df['StateHoliday'].fillna("0").astype(str)
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce').fillna(0)
df = df[df['Sales'] > 0].copy()   # remove zero sales rows if desired
df = df.sort_values(['Store','Date']).reset_index(drop=True)
print("After basic cleaning shape:", df.shape)

# -------------------------
# 4. Feature engineering
# -------------------------
def create_date_features(df):
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
    df['DayOfYear'] = df['Date'].dt.dayofyear
    df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
    df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
    df['DayOfWeek_sin'] = np.sin(2 * np.pi * df['DayOfWeek'] / 7)
    df['DayOfWeek_cos'] = np.cos(2 * np.pi * df['DayOfWeek'] / 7)
    df['IsWeekend'] = df['DayOfWeek'].isin([6,7]).astype(int)
    return df


df = create_date_features(df)

# lag & rolling (group by store)
df['Sales'] = df['Sales'].astype(float)
df['Sales_lag_1'] = df.groupby('Store')['Sales'].shift(1)
df['Sales_lag_7'] = df.groupby('Store')['Sales'].shift(7)
df['Sales_lag_14'] = df.groupby('Store')['Sales'].shift(14)

df['Rolling_mean_7'] = df.groupby('Store')['Sales'].shift(1).rolling(7, min_periods=1).mean().reset_index(level=0, drop=True)
df['Rolling_std_7'] = df.groupby('Store')['Sales'].shift(1).rolling(7, min_periods=1).std().reset_index(level=0, drop=True).fillna(0)
df['Expanding_mean'] = df.groupby('Store')['Sales'].apply(lambda x: x.shift(1).expanding().mean()).reset_index(level=0, drop=True).fillna(0)

# interactions
df['Promo_per_SchoolHoliday'] = df['Promo'] * df['SchoolHoliday']
df['Promo_per_Weekend'] = df['Promo'] * df['IsWeekend']
df['StateHoliday_enc'] = df['StateHoliday'].astype('category').cat.codes

# drop rows with missing lag features (first periods)
df = df.dropna(subset=['Sales_lag_1','Sales_lag_7']).reset_index(drop=True)
print("After features shape:", df.shape)

# -------------------------
# 5. Features & target
# -------------------------
FEATURES = [
    'Store','DayOfWeek','Customers','Promo','SchoolHoliday','StateHoliday_enc',
    'Year','Month','Day','WeekOfYear','DayOfYear','IsWeekend',
    'Sales_lag_1','Sales_lag_7','Sales_lag_14',
    'Rolling_mean_7','Rolling_std_7','Expanding_mean',
    'Promo_per_SchoolHoliday','Promo_per_Weekend',
    'Month_sin','Month_cos','DayOfWeek_sin','DayOfWeek_cos'
]
FEATURES = [c for c in FEATURES if c in df.columns]
TARGET = 'Sales'
print("Using features:", FEATURES)

# -------------------------
# 6. Chronological split: train / test
# -------------------------
df = df.sort_values('Date').reset_index(drop=True)
train_df = df[df['Date'] < DATE_SPLIT].copy()
test_df  = df[df['Date'] >= DATE_SPLIT].copy()
print("Train/Test shapes:", train_df.shape, test_df.shape)

X_train_full = train_df[FEATURES]
y_train_full_raw = train_df[TARGET].values
X_test = test_df[FEATURES]
y_test_raw = test_df[TARGET].values

# further split training to train/val chronologically (no shuffle)
split_idx = int(len(X_train_full) * 0.8)
X_train = X_train_full.iloc[:split_idx].copy()
y_train_raw = y_train_full_raw[:split_idx].copy()
X_val   = X_train_full.iloc[split_idx:].copy()
y_val_raw = y_train_full_raw[split_idx:].copy()
print("Train/Val/Test shapes:", X_train.shape, X_val.shape, X_test.shape)

# -------------------------
# 7. Transform target: log1p
# -------------------------
y_train = np.log1p(y_train_raw)
y_val   = np.log1p(y_val_raw)
y_test  = np.log1p(y_test_raw)  # optional for log metrics

# -------------------------
# 8. Train LightGBM on log-scale
# -------------------------
# LightGBM parameters
lgb_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'verbosity': -1,
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 127,
    'min_data_in_leaf': 200,
    'seed': RANDOM_STATE,
    'n_jobs': -1
}

print("Training LightGBM (with callbacks)...")
# instantiate
lgb_model = lgb.LGBMRegressor(**{k: v for k, v in lgb_params.items()})
# fit using callbacks early_stopping & silent logging
lgb_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    eval_metric='rmse',
    callbacks=[early_stopping(stopping_rounds=100), log_evaluation(period=0)]
)

# -------------------------
# 9. XGBoost training via xgb.train + DMatrix (robust across versions)
# -------------------------
print("Training XGBoost using xgb.train + DMatrix (with early stopping)...")

# prepare DMatrix for log-scale targets (pass DataFrame directly)
dtrain = xgb.DMatrix(X_train, label=y_train)
dval   = xgb.DMatrix(X_val,   label=y_val)
dtest  = xgb.DMatrix(X_test,  label=y_test)

# params for xgboost.train (we're training on log1p(Sales))
xgb_params = {
    'objective': 'reg:squarederror',
    'tree_method': 'hist',
    'eta': 0.05,
    'max_depth': 8,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'eval_metric': 'rmse',
    'seed': RANDOM_STATE,
    'verbosity': 0
}

num_boost_round = 1000
early_stopping_rounds = 100
evals = [(dtrain, 'train'), (dval, 'val')]

bst = xgb.train(
    params=xgb_params,
    dtrain=dtrain,
    num_boost_round=num_boost_round,
    evals=evals,
    early_stopping_rounds=early_stopping_rounds,
    verbose_eval=50
)

# robust best_iteration handling
best_iteration = getattr(bst, "best_iteration", None)
if best_iteration is None:
    best_iteration = getattr(bst, "best_ntree_limit", None)
if best_iteration is None:
    best_iteration = num_boost_round
print("XGBoost best_iteration:", best_iteration)

# -------------------------
# 10. Predictions (log-scale) - FIX: use iteration_range
# -------------------------
pred_xgb_val_log = bst.predict(dval, iteration_range=(0, best_iteration))
pred_xgb_test_log = bst.predict(dtest, iteration_range=(0, best_iteration))

# LightGBM predictions with num_iteration if available
num_it_lgb = getattr(lgb_model, 'best_iteration_', None)
pred_lgb_val_log = lgb_model.predict(X_val, num_iteration=num_it_lgb)
pred_lgb_test_log = lgb_model.predict(X_test, num_iteration=num_it_lgb)

# If CatBoost available, train and predict
if CATBOOST_AVAILABLE:
    print("Training CatBoost...")
    cat_model = CatBoostRegressor(iterations=800, learning_rate=0.05, depth=8, verbose=100, random_state=RANDOM_STATE)
    cat_model.fit(X_train, y_train, eval_set=(X_val, y_val), early_stopping_rounds=100)
    pred_cat_val_log = cat_model.predict(X_val)
    pred_cat_test_log = cat_model.predict(X_test)
else:
    cat_model = None

# back to real scale
pred_lgb_val = np.expm1(pred_lgb_val_log)
pred_xgb_val = np.expm1(pred_xgb_val_log)
if cat_model is not None:
    pred_cat_val = np.expm1(pred_cat_val_log)

pred_lgb_test = np.expm1(pred_lgb_test_log)
pred_xgb_test = np.expm1(pred_xgb_test_log)
if cat_model is not None:
    pred_cat_test = np.expm1(pred_cat_test_log)

# -------------------------
# 11. Compute validation RMSE and ensemble
# -------------------------
rmse_lgb_val = np.sqrt(mean_squared_error(y_val_raw, pred_lgb_val))
rmse_xgb_val = np.sqrt(mean_squared_error(y_val_raw, pred_xgb_val))
rmse_cat_val = np.sqrt(mean_squared_error(y_val_raw, pred_cat_val)) if cat_model is not None else np.inf

print(f"LGB val RMSE: {rmse_lgb_val:.2f}")
print(f"XGB val RMSE: {rmse_xgb_val:.2f}")
if cat_model is not None: print(f"Cat val RMSE: {rmse_cat_val:.2f}")

# weights = inverse RMSE
w_l = 1.0 / (rmse_lgb_val + 1e-9)
w_x = 1.0 / (rmse_xgb_val + 1e-9)
if cat_model is not None:
    w_c = 1.0 / (rmse_cat_val + 1e-9)

# ensemble validation
if cat_model is not None:
    ensemble_val = (w_l*pred_lgb_val + w_x*pred_xgb_val + w_c*pred_cat_val) / (w_l+w_x+w_c)
else:
    ensemble_val = (w_l*pred_lgb_val + w_x*pred_xgb_val) / (w_l + w_x)

rmse_ens_val = np.sqrt(mean_squared_error(y_val_raw, ensemble_val))
mae_ens_val = mean_absolute_error(y_val_raw, ensemble_val)
print(f"Ensemble (val) RMSE: {rmse_ens_val:.2f} | MAE: {mae_ens_val:.2f}")

# -------------------------
# 12. Evaluate on test set
# -------------------------
if cat_model is not None:
    pred_ens_test = (w_l*pred_lgb_test + w_x*pred_xgb_test + w_c*pred_cat_test) / (w_l + w_x + w_c)
else:
    pred_ens_test = (w_l*pred_lgb_test + w_x*pred_xgb_test) / (w_l + w_x)

rmse_test = np.sqrt(mean_squared_error(y_test_raw, pred_ens_test))
mae_test  = mean_absolute_error(y_test_raw, pred_ens_test)
print("\nFinal Test ensemble metrics (real scale):")
print(f"Ensemble Test RMSE: {rmse_test:.2f}")
print(f"Ensemble Test MAE : {mae_test:.2f}")

# -------------------------
# 13. Diagnostics plots
# -------------------------
sample_n = min(200, len(test_df))
plt.figure(figsize=(12,5))
plt.plot(test_df['Date'].values[:sample_n], test_df['Sales'].values[:sample_n], label='Actual', lw=1)
plt.plot(test_df['Date'].values[:sample_n], pred_ens_test[:sample_n], label='Predicted', lw=1)
plt.legend(); plt.title("Test: Actual vs Predicted (sample)"); plt.xticks(rotation=20); plt.show()

residuals = test_df['Sales'].values - pred_ens_test
plt.figure(figsize=(8,4)); sns.histplot(residuals, bins=80, kde=True); plt.title("Residuals (test)"); plt.show()

test_df = test_df.reset_index(drop=True)
test_df['predicted_sales'] = pred_ens_test
test_df['abs_error'] = np.abs(test_df['Sales'] - test_df['predicted_sales'])
err_by_day = test_df.groupby('DayOfWeek')['abs_error'].mean()
plt.figure(figsize=(8,4)); err_by_day.plot(kind='bar'); plt.title("Mean Abs Error by DayOfWeek (test)"); plt.show()
print("Mean abs error by DayOfWeek:\n", err_by_day)

# -------------------------
# 14. Learning curves and evals result plots
# -------------------------
# XGBoost evals_result
try:
    xgb_results = bst.evals_result()
    if 'train' in xgb_results and 'rmse' in xgb_results['train']:
        train_rmse = xgb_results['train']['rmse']
        val_rmse = xgb_results['val']['rmse']
        plt.figure(figsize=(10,5))
        plt.plot(train_rmse, label='xgb_train_rmse')
        plt.plot(val_rmse, label='xgb_val_rmse')
        plt.axvline(best_iteration, color='gray', linestyle='--', label='best_iter')
        plt.xlabel('boost round')
        plt.ylabel('rmse')
        plt.legend(); plt.title('XGBoost learning curve'); plt.show()
except Exception as e:
    print('Could not plot xgb evals_result:', e)

# LightGBM evals (if available through _recorded)
try:
    lgb_eval = getattr(lgb_model, 'evals_result_', None)
    if lgb_eval is not None and 'valid_0' in lgb_eval:
        l_train = lgb_eval['valid_0']['rmse'] if 'rmse' in lgb_eval['valid_0'] else None
        # LGB's recorded structure depends on versions; we attempt a simple plot if possible
        # Otherwise skip
except Exception as e:
    pass

# -------------------------
# 15. Feature importance & SHAP
# -------------------------
# XGBoost feature importance
try:
    fi = bst.get_score(importance_type='gain')
    fi_series = pd.Series(fi).sort_values(ascending=False)
    plt.figure(figsize=(8,6)); fi_series.head(30).plot(kind='bar'); plt.title('XGBoost feature importance (gain)'); plt.show()
except Exception as e:
    print('Could not compute XGB feature importance:', e)

# SHAP explanations (XGBoost)
if SHAP_AVAILABLE:
    try:
        # use TreeExplainer for XGBoost
        explainer = shap.TreeExplainer(bst)
        # shap expects numpy or DataFrame; use a sample to speed up
        sample_for_shap = X_val.sample(min(1000, len(X_val)), random_state=RANDOM_STATE)
        shap_values = explainer.shap_values(sample_for_shap)
        print('SHAP computed. Showing summary plot...')
        shap.summary_plot(shap_values, sample_for_shap, show=True)
        # save shap values for later use
        np.save(os.path.join(ARTIFACT_DIR, 'shap_vals_xgb.npy'), shap_values)
    except Exception as e:
        print('Failed to compute SHAP for XGBoost:', e)
else:
    print('SHAP not available; skipping SHAP explanations.')

# -------------------------
# 16. Save artifacts
# -------------------------
joblib.dump(lgb_model, os.path.join(ARTIFACT_DIR, "lgb_model_log1p.pkl"))
bst.save_model(os.path.join(ARTIFACT_DIR, "xgb_booster.json"))
if cat_model is not None:
    joblib.dump(cat_model, os.path.join(ARTIFACT_DIR, "cat_model_log1p.pkl"))
joblib.dump(FEATURES, os.path.join(ARTIFACT_DIR, "feature_list.pkl"))
meta = {
    "DATE_SPLIT": DATE_SPLIT,
    "features": FEATURES,
    "rmse_val": float(rmse_ens_val),
    "rmse_test": float(rmse_test)
}
with open(os.path.join(ARTIFACT_DIR, "meta.json"), "w") as f:
    json.dump(meta, f, indent=2)
print("Artifacts saved to:", ARTIFACT_DIR)

print("Pipeline finished. Next suggestions: (1) run Optuna tuning on LGB, (2) try per-store models/clustering, (3) produce SHAP explanations for top features more thoroughly.")
```

```python
# ---------- Optuna tuning for Colab (XGBoost & LightGBM) ----------
# Requirements (run once in Colab):
!pip install optuna xgboost lightgbm catboost --quiet

import os, time, json, joblib
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
import optuna

# detect GPU (Colab)
def detect_gpu():
    try:
        import subprocess, shlex
        out = subprocess.check_output(shlex.split("nvidia-smi -L"), stderr=subprocess.DEVNULL).decode().strip()
        return True if out else False
    except Exception:
        return False

GPU_AVAILABLE = detect_gpu()
print("GPU available:", GPU_AVAILABLE)

# safe defaults for Colab — you can increase N_TRIALS if you have more time/resources
N_TRIALS = 40    # default smaller for Colab; increase to 100-200 if you want longer search
NUM_BOOST_ROUND = 1500
EARLY_STOPPING = 80

# helper to evaluate RMSE on real scale given log1p preds
def rmse_real_scale(y_true_raw, y_pred_log1p):
    return np.sqrt(mean_squared_error(y_true_raw, np.expm1(y_pred_log1p)))

# XGBoost objective (uses gpu_hist if GPU_AVAILABLE)
def optuna_xgb_objective(trial):
    import xgboost as xgb
    param = {
        'objective': 'reg:squarederror',
        'eta': trial.suggest_loguniform('eta', 0.01, 0.2),
        'max_depth': trial.suggest_int('max_depth', 4, 12),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 20),
        'gamma': trial.suggest_float('gamma', 0.0, 5.0),
        'lambda': trial.suggest_loguniform('lambda', 1e-3, 10.0),
        'alpha': trial.suggest_loguniform('alpha', 1e-3, 10.0),
        'verbosity': 0,
        'seed': RANDOM_STATE
    }
    # enable GPU if present
    if GPU_AVAILABLE:
        param.update({'tree_method':'gpu_hist', 'predictor':'gpu_predictor'})

    dtrain_local = xgb.DMatrix(X_train, label=y_train)
    dval_local = xgb.DMatrix(X_val, label=y_val)
    evals = [(dtrain_local, 'train'), (dval_local, 'val')]

    bst_local = xgb.train(
        params=param,
        dtrain=dtrain_local,
        num_boost_round=NUM_BOOST_ROUND,
        evals=evals,
        early_stopping_rounds=EARLY_STOPPING,
        verbose_eval=False
    )

    best_iter = getattr(bst_local, 'best_iteration', None) or getattr(bst_local, 'best_ntree_limit', None) or NUM_BOOST_ROUND
    preds_val_log = bst_local.predict(dval_local, iteration_range=(0, best_iter))
    rmse_val = rmse_real_scale(y_val_raw, preds_val_log)
    trial.report(rmse_val, step=0)
    return rmse_val

# LightGBM objective (fast CPU/GPU depending on installation; we use CPU by default)
def optuna_lgb_objective(trial):
    import lightgbm as lgb
    param = {
        'objective': 'regression',
        'metric': 'rmse',
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.2),
        'num_leaves': trial.suggest_int('num_leaves', 31, 512),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 20, 1000),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.5, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.5, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 0, 10),
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-4, 10.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-4, 10.0),
        'seed': RANDOM_STATE,
        'n_jobs': -1,
        'verbosity': -1
    }

    lgb_train = lgb.Dataset(X_train, label=y_train)
    lgb_val = lgb.Dataset(X_val, label=y_val, reference=lgb_train)

    gbm = lgb.train(
        param,
        lgb_train,
        num_boost_round=NUM_BOOST_ROUND,
        valid_sets=[lgb_train, lgb_val],
        early_stopping_rounds=EARLY_STOPPING,
        verbose_eval=False
    )

    best_iter = gbm.best_iteration or NUM_BOOST_ROUND
    preds_val_log = gbm.predict(X_val, num_iteration=best_iter)
    rmse_val = rmse_real_scale(y_val_raw, preds_val_log)
    trial.report(rmse_val, step=0)
    return rmse_val

# Optuna study setup with pruning to save time
pruner = optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=0)
study_xgb = optuna.create_study(direction='minimize', study_name='xgb_colab', sampler=optuna.samplers.TPESampler(seed=RANDOM_STATE), pruner=pruner)
study_lgb = optuna.create_study(direction='minimize', study_name='lgb_colab', sampler=optuna.samplers.TPESampler(seed=RANDOM_STATE), pruner=pruner)

# Run studies (adjust n_trials as needed)
print("Starting XGBoost tuning...")
study_xgb.optimize(optuna_xgb_objective, n_trials=N_TRIALS, show_progress_bar=True)

print("Starting LightGBM tuning...")
study_lgb.optimize(optuna_lgb_objective, n_trials=N_TRIALS, show_progress_bar=True)

# Save studies
os.makedirs(ARTIFACT_DIR, exist_ok=True)
joblib.dump(study_xgb, os.path.join(ARTIFACT_DIR, "optuna_xgb_study_colab.pkl"))
joblib.dump(study_lgb, os.path.join(ARTIFACT_DIR, "optuna_lgb_study_colab.pkl"))
print("Studies saved to", ARTIFACT_DIR)
```

```python
# ===== Recreate preprocessing pipeline and save processed df =====
import pandas as pd, numpy as np, os, json
from datetime import datetime

DATA_PATH = "/content/drive/MyDrive/Rossmann%20Stores%20Data%20%282%29.csv"
ARTIFACT_DIR = "/content/drive/MyDrive/rossmann_artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

print("Loading raw CSV:", DATA_PATH)
df = pd.read_csv(DATA_PATH, parse_dates=['Date'], low_memory=False)
print("Raw shape:", df.shape)

# Basic cleaning (same as earlier pipeline)
df['Open'] = df.get('Open', pd.Series(1, index=df.index)).fillna(1).astype(int)
df = df[df['Open'] == 1].copy()
df['StateHoliday'] = df['StateHoliday'].fillna("0").astype(str)
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce').fillna(0)
# keep >0 as before (option: you may change later)
df = df[df['Sales'] > 0].copy()
df = df.sort_values(['Store','Date']).reset_index(drop=True)

# Date features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
df['DayOfYear'] = df['Date'].dt.dayofyear
df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
df['DayOfWeek_sin'] = np.sin(2 * np.pi * df['DayOfWeek'] / 7)
df['DayOfWeek_cos'] = np.cos(2 * np.pi * df['DayOfWeek'] / 7)
df['IsWeekend'] = df['DayOfWeek'].isin([6,7]).astype(int)

# Lags & rolling features
df['Sales_lag_1'] = df.groupby('Store')['Sales'].shift(1)
df['Sales_lag_7'] = df.groupby('Store')['Sales'].shift(7)
df['Sales_lag_14'] = df.groupby('Store')['Sales'].shift(14)

df['Rolling_mean_7'] = df.groupby('Store')['Sales'].shift(1).rolling(7, min_periods=1).mean().reset_index(level=0, drop=True)
df['Rolling_std_7'] = df.groupby('Store')['Sales'].shift(1).rolling(7, min_periods=1).std().reset_index(level=0, drop=True).fillna(0)
df['Expanding_mean'] = df.groupby('Store')['Sales'].apply(lambda x: x.shift(1).expanding().mean()).reset_index(level=0, drop=True).fillna(0)

# interactions & encodings
df['Promo_per_SchoolHoliday'] = df.get('Promo',0) * df.get('SchoolHoliday',0)
df['Promo_per_Weekend'] = df.get('Promo',0) * df['IsWeekend']
df['StateHoliday_enc'] = df['StateHoliday'].astype('category').cat.codes

# Drop rows missing lag features (as before)
df = df.dropna(subset=['Sales_lag_1','Sales_lag_7']).reset_index(drop=True)

# Save processed dataframe (so you don't redo this every restart)
proc_path = os.path.join(ARTIFACT_DIR, "rossmann_preprocessed.csv")
df.to_csv(proc_path, index=False)
print("Saved preprocessed df to:", proc_path)
print("Processed shape:", df.shape)
```

```python
# ===== Add MA + per-store target encoding, rebuild datasets, quick retrain LGB & XGB =====
import os, joblib, json, warnings
warnings.filterwarnings("ignore")
import pandas as pd, numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
import xgboost as xgb
import lightgbm as lgb

ARTIFACT_DIR = "/content/drive/MyDrive/rossmann_artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

# --- 0. Safety checks: ensure df exists and Date is datetime ---
if 'df' not in globals():
    raise RuntimeError("DataFrame 'df' not found in memory. Please run the preprocessing cells (load CSV, feature engineering) first.")
if not np.issubdtype(df['Date'].dtype, np.datetime64):
    df['Date'] = pd.to_datetime(df['Date'])

# make a working copy sorted by store/date
df = df.sort_values(['Store','Date']).reset_index(drop=True)

# --- 1. Add moving averages (shifted to avoid leakage) ---
def add_store_mas(df, windows=[7,30,90,180,365], group_col='Store', target='Sales'):
    for w in windows:
        col = f"{target}_ma_{w}"
        df[col] = df.groupby(group_col)[target].shift(1).rolling(window=w, min_periods=1).mean().reset_index(level=0, drop=True)
    return df

df = add_store_mas(df, windows=[7,30,90,180,365])

# --- 2. Per-store stats from train split only (avoid leakage) ---
DATE_SPLIT = "2015-06-01"  # use your existing DATE_SPLIT
train_df = df[df['Date'] < DATE_SPLIT].copy()
test_df  = df[df['Date'] >= DATE_SPLIT].copy()

store_stats = train_df.groupby('Store')['Sales'].agg([
    ('store_mean_sales','mean'),
    ('store_median_sales','median'),
    ('store_std_sales','std'),
    ('store_count_days','count')
]).fillna(0)

# promo rate if exists
if 'Promo' in train_df.columns:
    promo_rate = train_df.groupby('Store')['Promo'].mean().rename('store_promo_rate')
    store_stats = store_stats.join(promo_rate, how='left')

# map stats to full df
for c in store_stats.columns:
    df[c] = df['Store'].map(store_stats[c]).fillna(store_stats[c].mean())

# --- 3. Target encoding (smooth) computed from train only ---
def smooth_target_encoding(train_series, train_target, full_series, smoothing=10):
    # train_series & train_target are Series aligned
    agg = train_target.groupby(train_series).agg(['mean','count'])
    prior = train_target.mean()
    smooth = (agg['count'] * agg['mean'] + smoothing * prior) / (agg['count'] + smoothing)
    return full_series.map(smooth).fillna(prior)

df['store_te_sales'] = smooth_target_encoding(train_df['Store'], train_df['Sales'], df['Store'], smoothing=10)

# --- 4. Recreate features list and splits (ensure original features kept) ---
# reuse your previous FEATURES list if exists, else construct
desired_features = [
    'Store','DayOfWeek','Customers','Promo','SchoolHoliday','StateHoliday_enc',
    'Year','Month','Day','WeekOfYear','DayOfYear','IsWeekend',
    'Sales_lag_1','Sales_lag_7','Sales_lag_14',
    'Rolling_mean_7','Rolling_std_7','Expanding_mean',
    'Promo_per_SchoolHoliday','Promo_per_Weekend',
    'Month_sin','Month_cos','DayOfWeek_sin','DayOfWeek_cos'
]
# add the new engineered columns
extra_feats = ['Sales_ma_7','Sales_ma_30','Sales_ma_90','Sales_ma_180','Sales_ma_365',
               'store_mean_sales','store_median_sales','store_std_sales','store_promo_rate','store_te_sales']
FEATURES = [c for c in (desired_features + extra_feats) if c in df.columns]
TARGET = 'Sales'
print("Final features used:", FEATURES)

# drop rows where required lag features missing (same as before)
df = df.sort_values('Date').reset_index(drop=True)
df = df.dropna(subset=['Sales_lag_1','Sales_lag_7']).reset_index(drop=True)

# rebuild splits
train_df = df[df['Date'] < DATE_SPLIT].copy()
test_df  = df[df['Date'] >= DATE_SPLIT].copy()

# within train, chronological split for val
split_idx = int(len(train_df) * 0.8)
train_part = train_df.iloc[:split_idx].copy()
val_part = train_df.iloc[split_idx:].copy()

X_train = train_part[FEATURES]
y_train_raw = train_part[TARGET].values
X_val   = val_part[FEATURES]
y_val_raw = val_part[TARGET].values
X_test  = test_df[FEATURES]
y_test_raw = test_df[TARGET].values

# transform targets (log1p) as before
y_train = np.log1p(y_train_raw)
y_val   = np.log1p(y_val_raw)
y_test  = np.log1p(y_test_raw)

print("Shapes -> X_train, X_val, X_test:", X_train.shape, X_val.shape, X_test.shape)

# --- 5. Try to load best LightGBM params from saved Optuna study (if present) ---
best_lgb_params = None
study_path = os.path.join(ARTIFACT_DIR, "optuna_lgb_study_final.pkl")
if os.path.exists(study_path):
    try:
        study = joblib.load(study_path)
        best_lgb_params = study.best_trial.params
        print("Loaded best LGB params from study:", best_lgb_params)
    except Exception as e:
        print("Failed to load study params:", e)

# adapt params into LightGBM format (if found)
if best_lgb_params is not None:
    lgb_params = {
        'objective':'regression','metric':'rmse','boosting_type':'gbdt',
        'learning_rate': best_lgb_params.get('learning_rate', 0.05),
        'num_leaves': int(best_lgb_params.get('num_leaves', 127)),
        'min_data_in_leaf': int(best_lgb_params.get('min_data_in_leaf', 20)),
        'feature_fraction': best_lgb_params.get('feature_fraction', 0.8),
        'bagging_fraction': best_lgb_params.get('bagging_fraction', 0.8),
        'bagging_freq': int(best_lgb_params.get('bagging_freq', 0)),
        'lambda_l1': best_lgb_params.get('lambda_l1', 0.0),
        'lambda_l2': best_lgb_params.get('lambda_l2', 0.0),
        'verbosity': -1, 'seed': 42, 'n_jobs': -1
    }
else:
    lgb_params = {'objective':'regression','metric':'rmse','learning_rate':0.05,'num_leaves':127,'min_data_in_leaf':50,'verbosity':-1,'n_jobs':-1,'seed':42}

# --- 6. Quick train LightGBM (on log1p target) ---
lgb_train = lgb.Dataset(X_train, label=y_train)
lgb_val = lgb.Dataset(X_val, label=y_val, reference=lgb_train)
gbm = lgb.train(
    lgb_params, lgb_train,
    num_boost_round=1500,
    valid_sets=[lgb_train, lgb_val],
    callbacks=[lgb.early_stopping(stopping_rounds=80), lgb.log_evaluation(period=0)]
)
best_iter_lgb = getattr(gbm, 'best_iteration', 1500)
pred_lgb_val_log = gbm.predict(X_val, num_iteration=best_iter_lgb)
pred_lgb_test_log = gbm.predict(X_test, num_iteration=best_iter_lgb)
pred_lgb_val = np.expm1(pred_lgb_val_log)
pred_lgb_test = np.expm1(pred_lgb_test_log)

rmse_lgb_val = np.sqrt(mean_squared_error(y_val_raw, pred_lgb_val))
rmse_lgb_test = np.sqrt(mean_squared_error(y_test_raw, pred_lgb_test))
print(f"LGB val RMSE: {rmse_lgb_val:.2f} | test RMSE: {rmse_lgb_test:.2f}")

# --- 7. Quick train XGBoost (use simple params or load best if you have) ---
# try load XGB best params from earlier study if exists
best_xgb_params = None
xgb_study_path = os.path.join(ARTIFACT_DIR, "optuna_xgb_study.pkl")
if os.path.exists(xgb_study_path):
    try:
        stx = joblib.load(xgb_study_path)
        best_xgb_params = stx.best_trial.params
        print("Loaded best XGB params from optuna_xgb_study.pkl")
    except Exception:
        best_xgb_params = None

# default xgb params
xgb_params = {
    'objective':'reg:squarederror','eta':0.05,'max_depth':8,'subsample':0.8,'colsample_bytree':0.8,'seed':42,'verbosity':0
}
if best_xgb_params is not None:
    # map fields conservatively
    xgb_params.update({k: best_xgb_params[k] for k in ['eta','max_depth','subsample','colsample_bytree'] if k in best_xgb_params})

dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)
dtest = xgb.DMatrix(X_test, label=y_test)

bst = xgb.train(xgb_params, dtrain, num_boost_round=1000, evals=[(dtrain,'train'),(dval,'val')],
                early_stopping_rounds=80, verbose_eval=False)
best_iter_xgb = getattr(bst, 'best_iteration', 1000)
pred_xgb_val_log = bst.predict(dval, iteration_range=(0,best_iter_xgb))
pred_xgb_test_log = bst.predict(dtest, iteration_range=(0,best_iter_xgb))
pred_xgb_val = np.expm1(pred_xgb_val_log)
pred_xgb_test = np.expm1(pred_xgb_test_log)

rmse_xgb_val = np.sqrt(mean_squared_error(y_val_raw, pred_xgb_val))
rmse_xgb_test = np.sqrt(mean_squared_error(y_test_raw, pred_xgb_test))
print(f"XGB val RMSE: {rmse_xgb_val:.2f} | test RMSE: {rmse_xgb_test:.2f}")

# --- 8. Simple weighted ensemble (weights inverse RMSE on val) ---
w_l = 1.0 / (rmse_lgb_val + 1e-9)
w_x = 1.0 / (rmse_xgb_val + 1e-9)
pred_ens_val = (w_l*pred_lgb_val + w_x*pred_xgb_val) / (w_l + w_x)
pred_ens_test = (w_l*pred_lgb_test + w_x*pred_xgb_test) / (w_l + w_x)

rmse_ens_val = np.sqrt(mean_squared_error(y_val_raw, pred_ens_val))
rmse_ens_test = np.sqrt(mean_squared_error(y_test_raw, pred_ens_test))
mae_ens_test = mean_absolute_error(y_test_raw, pred_ens_test)
print(f"Ensemble val RMSE: {rmse_ens_val:.2f} | Ensemble test RMSE: {rmse_ens_test:.2f} | MAE: {mae_ens_test:.2f}")

# --- 9. Save models & feature list & metrics ---
joblib.dump(gbm, os.path.join(ARTIFACT_DIR, "lgb_gbm_ma_tuned.pkl"))
bst.save_model(os.path.join(ARTIFACT_DIR, "xgb_bst_ma_tuned.json"))
joblib.dump(FEATURES, os.path.join(ARTIFACT_DIR, "feature_list_with_ma.pkl"))

metrics = {
    "rmse_lgb_val": float(rmse_lgb_val),
    "rmse_xgb_val": float(rmse_xgb_val),
    "rmse_ens_val": float(rmse_ens_val),
    "rmse_ens_test": float(rmse_ens_test),
    "mae_ens_test": float(mae_ens_test),
    "best_iter_lgb": int(best_iter_lgb),
    "best_iter_xgb": int(best_iter_xgb)
}
with open(os.path.join(ARTIFACT_DIR, "ma_tuned_metrics.json"), "w") as f:
    json.dump(metrics, f, indent=2)

print("Saved models & metrics to:", ARTIFACT_DIR)
```

```python
# ===== Per-store clustering + per-cluster LGB models =====
import pandas as pd, numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import joblib, os

ARTIFACT_DIR = "/content/drive/MyDrive/rossmann_artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

# 1. compute store-level stats (use train_df)
DATE_SPLIT = "2015-06-01"
train_df_all = df[df['Date'] < DATE_SPLIT].copy()

store_feats = train_df_all.groupby('Store').agg({
    'Sales': ['mean','std','skew'],
    'Promo': ['mean'],
    'Customers': ['mean']
})
store_feats.columns = ['_'.join(col).strip() for col in store_feats.columns.values]
store_feats = store_feats.fillna(0)

# 2. scale and cluster
scaler = StandardScaler()
S = scaler.fit_transform(store_feats.values)
k = 8  # start with 6-12 clusters; tune later
km = KMeans(n_clusters=k, random_state=42, n_init=20)
clusters = km.fit_predict(S)
store_feats['cluster'] = clusters
# map to df
cluster_map = store_feats['cluster'].to_dict()
df['store_cluster'] = df['Store'].map(cluster_map).fillna(-1).astype(int)

# Save artifacts
joblib.dump(km, os.path.join(ARTIFACT_DIR, "store_kmeans.pkl"))
joblib.dump(scaler, os.path.join(ARTIFACT_DIR, "store_cluster_scaler.pkl"))
print("Assigned clusters, sample:\n", df[['Store','store_cluster']].drop_duplicates().head())

# 3. Train one LGB per cluster (fast loop)
import lightgbm as lgb
cluster_models = {}
for c in sorted(df['store_cluster'].unique()):
    if c < 0: continue
    mask_train = (df['Date'] < DATE_SPLIT) & (df['store_cluster']==c)
    part = df[mask_train].copy()
    # chronological split into train/val inside cluster (80/20)
    part = part.sort_values('Date')
    split_idx = int(len(part)*0.8)
    if split_idx < 100:
        print(f"Cluster {c} too small, skipping separate model.")
        continue
    tr = part.iloc[:split_idx]
    va = part.iloc[split_idx:]
    X_tr = tr[FEATURES]; y_tr = np.log1p(tr['Sales'].values)
    X_va = va[FEATURES]; y_va = np.log1p(va['Sales'].values)

    lgb_train = lgb.Dataset(X_tr, label=y_tr)
    lgb_val = lgb.Dataset(X_va, label=y_va, reference=lgb_train)
    params = {'objective':'regression','metric':'rmse','learning_rate':0.05,'num_leaves':127,'verbosity':-1,'seed':42,'n_jobs':-1}
    bst = lgb.train(params, lgb_train, num_boost_round=1000, valid_sets=[lgb_train,lgb_val], callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)])
    cluster_models[c] = bst
    print(f"Trained LGB for cluster {c} | best_iter {bst.best_iteration}")

# save cluster models
joblib.dump(cluster_models, os.path.join(ARTIFACT_DIR,"cluster_models.pkl"))
print("Saved cluster models.")
```

```python
# ===== Stacking with TimeSeriesSplit (rolling) =====
from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import Ridge
import numpy as np

# helper: time-series folds on train_df (chronological)
train_df = df[df['Date'] < DATE_SPLIT].sort_values('Date').reset_index(drop=True)
n_splits = 5
tscv = TimeSeriesSplit(n_splits=n_splits)

base_models = []
base_preds_val = np.zeros((len(train_df), 5))  # placeholder if 5 base models, adjust dynamically
base_val_list = []  # collect per model

# Define base model factories
def train_xgb_on_fold(X_tr, y_tr, X_va, params_override=None):
    dtr = xgb.DMatrix(X_tr, label=y_tr)
    dva = xgb.DMatrix(X_va)
    params = {'objective':'reg:squarederror','eta':0.05,'max_depth':8,'subsample':0.8,'colsample_bytree':0.8,'verbosity':0,'seed':42}
    if params_override: params.update(params_override)
    bst = xgb.train(params, dtr, num_boost_round=1000, evals=[(dtr,'train')], early_stopping_rounds=80, verbose_eval=False)
    return bst

# Example: create 3 XGB variants + 1 LGB
xgb_variants = [
    {'eta':0.05,'max_depth':8},
    {'eta':0.03,'max_depth':10},
    {'eta':0.07,'max_depth':6}
]

# Build OOF preds for stacking
oof_preds = {f"xgb_{i}": np.zeros(len(train_df)) for i in range(len(xgb_variants))}
oof_preds['lgb'] = np.zeros(len(train_df))

for i, (train_index, val_index) in enumerate(tscv.split(train_df)):
    tr = train_df.iloc[train_index]; va = train_df.iloc[val_index]
    X_tr = tr[FEATURES]; y_tr = np.log1p(tr['Sales'].values)
    X_va = va[FEATURES]; y_va = np.log1p(va['Sales'].values)

    # train each xgb variant and fill OOF for this fold
    for j, pv in enumerate(xgb_variants):
        bst = train_xgb_on_fold(X_tr, y_tr, X_va, params_override=pv)
        preds_va = bst.predict(xgb.DMatrix(X_va), iteration_range=(0, getattr(bst,'best_iteration',1000)))
        oof_preds[f"xgb_{j}"][val_index] = preds_va

    # also train lightgbm on this fold
    dtr = lgb.Dataset(X_tr, label=y_tr)
    dval = lgb.Dataset(X_va, label=y_va, reference=dtr)
    lparams = {'objective':'regression','metric':'rmse','learning_rate':0.05,'num_leaves':127,'verbosity':-1,'seed':42}
    lbst = lgb.train(lparams, dtr, num_boost_round=1000, valid_sets=[dtr,dval], callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)])
    preds_lgb = lbst.predict(X_va, num_iteration=getattr(lbst,'best_iteration',1000))
    oof_preds['lgb'][val_index] = preds_lgb

print("Built OOF preds for stacking.")

# Fit meta-learner (Ridge) on OOF predictions
meta_X = np.vstack([oof_preds[k] for k in sorted(oof_preds.keys())]).T
meta_y = np.log1p(train_df['Sales'].values)
meta_model = Ridge(alpha=1.0)
meta_model.fit(meta_X, meta_y)
joblib.dump(meta_model, os.path.join(ARTIFACT_DIR, "meta_ridge.pkl"))
print("Trained meta-learner and saved.")
```

```python
# ===== Entity embeddings with Keras (Store + DayOfWeek) =====
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# prepare dataset (train only)
train_df_full = df[df['Date'] < DATE_SPLIT].sort_values('Date').reset_index(drop=True)
# ensure indices contiguous
cat_store = train_df_full['Store'].astype('int').values
cat_dow = train_df_full['DayOfWeek'].astype('int').values
# numeric features
num_feats = ['Customers','Promo','Rolling_mean_7','Sales_lag_7']  # extend as desired
X_num = train_df_full[num_feats].fillna(0).values
y_nn = np.log1p(train_df_full['Sales'].values)

# model hyperparams
n_stores = int(df['Store'].nunique()) + 1
embed_dim = min(50, max(8, n_stores//20))

# build model
input_store = layers.Input(shape=(1,), name='store_input')
input_dow = layers.Input(shape=(1,), name='dow_input')
input_num = layers.Input(shape=(X_num.shape[1],), name='num_input')

emb_store = layers.Embedding(input_dim=n_stores, output_dim=embed_dim, name='store_emb')(input_store)
emb_store = layers.Flatten()(emb_store)
emb_dow = layers.Embedding(input_dim=8, output_dim=4, name='dow_emb')(input_dow)
emb_dow = layers.Flatten()(emb_dow)

x = layers.Concatenate()([emb_store, emb_dow, input_num])
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(128, activation='relu')(x)
output = layers.Dense(1, activation='linear')(x)

nn_model = keras.Model(inputs=[input_store, input_dow, input_num], outputs=output)
nn_model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3), loss='mse')

# train (use validation split)
history = nn_model.fit(
    {'store_input':cat_store, 'dow_input':cat_dow, 'num_input':X_num},
    y_nn,
    validation_split=0.1,
    epochs=12, batch_size=1024, verbose=2
)

# Save model and extract embeddings (store)
nn_model.save(os.path.join(ARTIFACT_DIR, "nn_embedding_model.keras"))

store_embedding_layer = nn_model.get_layer('store_emb')
emb_weights = store_embedding_layer.get_weights()[0]  # shape (n_stores, embed_dim)
# create map: store -> vector (as list) and add to df as new features via dimension reduction or PCA
import pandas as pd
emb_df = pd.DataFrame(emb_weights, index=range(n_stores)).reset_index().rename(columns={'index':'Store'})
# join small PCA projection for each store to df
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
emb_proj = pca.fit_transform(emb_weights)
for i in range(emb_proj.shape[1]):
    df[f'store_emb_pca_{i}'] = df['Store'].map(dict(zip(range(n_stores), emb_proj[:,i])))
# Save embeddings/proj
joblib.dump({'emb_weights':emb_weights, 'pca':pca}, os.path.join(ARTIFACT_DIR, "store_embeddings.pkl"))
print("Saved NN embeddings and PCA projection. New features added to df: ", [f'store_emb_pca_{i}' for i in range(emb_proj.shape[1])])
```

```python
# ====== Save NN, extract store embeddings, add to df, retrain GBMs + stacking ======
import os, joblib, json, warnings, numpy as np, pandas as pd
warnings.filterwarnings("ignore")
ARTIFACT_DIR = "/content/drive/MyDrive/rossmann_artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

# 1) Save NN model properly (use .keras)
try:
    nn_save_path = os.path.join(ARTIFACT_DIR, "nn_embedding_model.keras")
    nn_model.save(nn_save_path)
    print("Saved NN model to:", nn_save_path)
except Exception as e:
    print("Failed to save nn_model:", e)
    # attempt h5
    try:
        nn_save_path = os.path.join(ARTIFACT_DIR, "nn_embedding_model.h5")
        nn_model.save(nn_save_path)
        print("Saved NN model (h5) to:", nn_save_path)
    except Exception as e2:
        raise RuntimeError("Unable to save NN model: " + str(e2))

# 2) Extract store embedding weights
from tensorflow.keras.models import load_model
try:
    # load back to ensure consistent API
    loaded_nn = load_model(nn_save_path)
    print("Loaded nn model for extraction.")
except Exception:
    loaded_nn = nn_model  # fallback

# try locate embedding layer named 'store_emb' or find first Embedding layer
emb_layer = None
for layer in loaded_nn.layers:
    if 'emb' in layer.name and 'store' in layer.name:
        emb_layer = layer
        break
if emb_layer is None:
    # find first Embedding layer
    for layer in loaded_nn.layers:
        if layer.__class__.__name__ == 'Embedding':
            emb_layer = layer
            break

if emb_layer is None:
    raise RuntimeError("No embedding layer found in the NN. Check layer names.")

emb_weights = emb_layer.get_weights()[0]  # shape: (n_stores, embed_dim)
print("Embedding weights shape:", emb_weights.shape)

# 3) PCA/DimReduce embeddings -> add k components as features
from sklearn.decomposition import PCA
pca_n = 8 if emb_weights.shape[1] >= 8 else emb_weights.shape[1]
pca = PCA(n_components=pca_n, random_state=42)
emb_proj = pca.fit_transform(emb_weights)  # shape (n_stores, pca_n)

# map embedding projections to df (Store indices assumed 0..n_stores-1 or matching embedding index)
# If your Store IDs are 1..N you may need to shift indices; detect that:
store_ids = np.arange(emb_weights.shape[0])
# create mapping; assume embedding row index == Store id (if not, please adjust mapping accordingly)
emb_map = {int(s): emb_proj[i] for i, s in enumerate(store_ids)}

# If your df['Store'] indices start at 1, and embedding index started at 0, check and adjust:
min_store_in_df = int(df['Store'].min())
if min_store_in_df == 1 and list(emb_map.keys())[0] == 0:
    # shift keys by +1
    emb_map = {k+1: v for k,v in emb_map.items()}

# add columns to df
for i in range(emb_proj.shape[1]):
    col = f'store_emb_pca_{i}'
    df[col] = df['Store'].map(lambda s: emb_map.get(int(s), np.zeros(emb_proj.shape[1]))[i]).fillna(0)

# save embeddings and pca
joblib.dump({'emb_weights':emb_weights,'pca':pca,'emb_map':emb_map}, os.path.join(ARTIFACT_DIR,"store_embeddings_and_pca.pkl"))
print("Saved store embeddings + pca to artifacts.")

# 4) Rebuild FEATURES to include new embedding features
embed_cols = [f'store_emb_pca_{i}' for i in range(emb_proj.shape[1])]
# ensure FEATURES var exists; otherwise reconstruct conservative list
try:
    FEATURES
except NameError:
    # fallback: pick numeric-like columns we used before
    FEATURES = [c for c in ['Store','DayOfWeek','Customers','Promo','SchoolHoliday','StateHoliday_enc',
                           'Year','Month','Day','WeekOfYear','DayOfYear','IsWeekend',
                           'Sales_lag_1','Sales_lag_7','Sales_lag_14',
                           'Rolling_mean_7','Rolling_std_7','Expanding_mean'] if c in df.columns]

# extend features
for c in embed_cols:
    if c not in FEATURES:
        FEATURES.append(c)
print("New FEATURES length:", len(FEATURES))

# 5) Recreate train/val/test splits and matrices (chronological)
DATE_SPLIT = globals().get('DATE_SPLIT', "2015-06-01")
df = df.sort_values('Date').reset_index(drop=True)
df = df.dropna(subset=['Sales_lag_1','Sales_lag_7']).reset_index(drop=True)

train_df = df[df['Date'] < DATE_SPLIT].copy()
test_df  = df[df['Date'] >= DATE_SPLIT].copy()
split_idx = int(len(train_df) * 0.8)
train_part = train_df.iloc[:split_idx].copy()
val_part = train_df.iloc[split_idx:].copy()

X_train = train_part[FEATURES]; y_train_raw = train_part['Sales'].values
X_val   = val_part[FEATURES];  y_val_raw   = val_part['Sales'].values
X_test  = test_df[FEATURES];   y_test_raw  = test_df['Sales'].values

y_train = np.log1p(y_train_raw)
y_val   = np.log1p(y_val_raw)
y_test  = np.log1p(y_test_raw)

print("Shapes ->", X_train.shape, X_val.shape, X_test.shape)

# 6) Retrain LightGBM & XGBoost quickly using new features (final models)
import lightgbm as lgb, xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error

# LightGBM params (use tuned params if you have)
lgb_params = globals().get('lgb_params', {'objective':'regression','metric':'rmse','learning_rate':0.05,'num_leaves':127,'min_data_in_leaf':50,'verbosity':-1,'n_jobs':-1,'seed':42})
lgb_train = lgb.Dataset(X_train, label=y_train)
lgb_val = lgb.Dataset(X_val, label=y_val, reference=lgb_train)
gbm_new = lgb.train(lgb_params, lgb_train, num_boost_round=1500, valid_sets=[lgb_train,lgb_val],
                    callbacks=[lgb.early_stopping(stopping_rounds=80), lgb.log_evaluation(period=0)])
best_iter_lgb = getattr(gbm_new,'best_iteration',1500)
pred_lgb_val_log = gbm_new.predict(X_val, num_iteration=best_iter_lgb)
pred_lgb_test_log = gbm_new.predict(X_test, num_iteration=best_iter_lgb)
pred_lgb_val = np.expm1(pred_lgb_val_log); pred_lgb_test = np.expm1(pred_lgb_test_log)
rmse_lgb_val = np.sqrt(mean_squared_error(y_val_raw, pred_lgb_val)); rmse_lgb_test = np.sqrt(mean_squared_error(y_test_raw, pred_lgb_test))
print(f"New LGB val RMSE: {rmse_lgb_val:.2f} | test RMSE: {rmse_lgb_test:.2f}")

# XGBoost quick retrain
xgb_params = globals().get('xgb_params', {'objective':'reg:squarederror','eta':0.05,'max_depth':8,'subsample':0.8,'colsample_bytree':0.8,'verbosity':0,'seed':42})
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)
dtest = xgb.DMatrix(X_test, label=y_test)
bst_new = xgb.train(xgb_params, dtrain, num_boost_round=1000, evals=[(dtrain,'train'),(dval,'val')], early_stopping_rounds=80, verbose_eval=False)
best_iter_xgb = getattr(bst_new, 'best_iteration', 1000)
pred_xgb_val_log = bst_new.predict(dval, iteration_range=(0,best_iter_xgb))
pred_xgb_test_log = bst_new.predict(dtest, iteration_range=(0,best_iter_xgb))
pred_xgb_val = np.expm1(pred_xgb_val_log); pred_xgb_test = np.expm1(pred_xgb_test_log)
rmse_xgb_val = np.sqrt(mean_squared_error(y_val_raw, pred_xgb_val)); rmse_xgb_test = np.sqrt(mean_squared_error(y_test_raw, pred_xgb_test))
print(f"New XGB val RMSE: {rmse_xgb_val:.2f} | test RMSE: {rmse_xgb_test:.2f}")

# 7) Simple ensemble (inv-RMSE weights) and save models
w_l = 1.0/(rmse_lgb_val+1e-9); w_x = 1.0/(rmse_xgb_val+1e-9)
pred_ens_val = (w_l*pred_lgb_val + w_x*pred_xgb_val)/(w_l+w_x)
pred_ens_test = (w_l*pred_lgb_test + w_x*pred_xgb_test)/(w_l+w_x)
rmse_ens_val = np.sqrt(mean_squared_error(y_val_raw, pred_ens_val)); rmse_ens_test = np.sqrt(mean_squared_error(y_test_raw, pred_ens_test))
mae_ens_test = mean_absolute_error(y_test_raw, pred_ens_test)
print(f"Ensemble val RMSE: {rmse_ens_val:.2f} | Ensemble test RMSE: {rmse_ens_test:.2f} | MAE: {mae_ens_test:.2f}")

# Save models & feature list
joblib.dump(gbm_new, os.path.join(ARTIFACT_DIR, "lgb_with_emb.pkl"))
bst_new.save_model(os.path.join(ARTIFACT_DIR, "xgb_with_emb.json"))
joblib.dump(FEATURES, os.path.join(ARTIFACT_DIR, "features_with_emb.pkl"))
metrics = {"rmse_lgb_val":float(rmse_lgb_val),"rmse_xgb_val":float(rmse_xgb_val),
           "rmse_ens_val":float(rmse_ens_val),"rmse_ens_test":float(rmse_ens_test),"mae_ens_test":float(mae_ens_test)}
with open(os.path.join(ARTIFACT_DIR,"with_embeddings_metrics.json"), "w") as f:
    json.dump(metrics, f, indent=2)
print("Saved models and metrics to ARTIFACT_DIR.")
```

```python
# ===== Repair FEATURES / add missing engineered features automatically =====
import os, joblib, pandas as pd, numpy as np
from sklearn.decomposition import PCA

ARTIFACT_DIR = "/content/drive/MyDrive/rossmann_artifacts"
proc_path = os.path.join(ARTIFACT_DIR, "rossmann_preprocessed.csv")
feat_path = os.path.join(ARTIFACT_DIR, "features_with_emb.pkl")
emb_path = os.path.join(ARTIFACT_DIR, "store_embeddings_and_pca.pkl")

# 1) load processed df
print("Loading processed df...")
df = pd.read_csv(proc_path, parse_dates=['Date'], low_memory=False)
print("Loaded df with shape:", df.shape)

# 2) load FEATURES (if exists) otherwise fall back to known lists
try:
    FEATURES = joblib.load(feat_path)
    print("Loaded FEATURES from:", feat_path)
except Exception as e:
    print("Could not load features_with_emb.pkl:", e)
    try:
        FEATURES = joblib.load(os.path.join(ARTIFACT_DIR, "feature_list_with_ma.pkl"))
        print("Loaded fallback feature_list_with_ma.pkl")
    except Exception as e2:
        print("Fallback also failed:", e2)
        FEATURES = [
            'Store','DayOfWeek','Customers','Promo','SchoolHoliday','StateHoliday_enc',
            'Year','Month','Day','WeekOfYear','DayOfYear','IsWeekend',
            'Sales_lag_1','Sales_lag_7','Sales_lag_14','Rolling_mean_7','Rolling_std_7',
            'Expanding_mean','Promo_per_SchoolHoliday','Promo_per_Weekend',
            'Month_sin','Month_cos','DayOfWeek_sin','DayOfWeek_cos'
        ]
        print("Using conservative default FEATURES.")

print("Initial FEATURES count:", len(FEATURES))

# 3) show df columns and compute intersection
df_cols = set(df.columns.tolist())
feat_set = list(FEATURES)
present = [c for c in feat_set if c in df_cols]
missing = [c for c in feat_set if c not in df_cols]

print(f"Columns in df: {len(df_cols)} | FEATURES present: {len(present)} | missing: {len(missing)}")
if missing:
    print("Missing features samples (first 30):", missing[:30])

# 4) Create common missing engineered features if needed
# Moving averages & store stats & target encoding
created = []
# moving averages names
ma_names = ['Sales_ma_7','Sales_ma_30','Sales_ma_90','Sales_ma_180','Sales_ma_365']
for col in ma_names:
    if col in missing:
        w = int(col.split('_')[-1])
        print("Creating", col)
        df[col] = df.sort_values(['Store','Date']).groupby('Store')['Sales'].shift(1).rolling(window=w, min_periods=1).mean().reset_index(level=0, drop=True)
        created.append(col)

# store stats from existing preprocessed (compute from training part if Date split exists)
if any(x in missing for x in ['store_mean_sales','store_median_sales','store_std_sales','store_promo_rate','store_te_sales']):
    print("Computing per-store stats from data (train split assumption)...")
    DATE_SPLIT = globals().get('DATE_SPLIT', "2015-06-01")
    train_df = df[df['Date'] < DATE_SPLIT].copy()
    sagg = train_df.groupby('Store')['Sales'].agg(['mean','median','std','count']).rename(columns={'mean':'store_mean_sales','median':'store_median_sales','std':'store_std_sales','count':'store_count_days'})
    # promo rate
    if 'Promo' in df.columns:
        promo_rate = train_df.groupby('Store')['Promo'].mean().rename('store_promo_rate')
        sagg = sagg.join(promo_rate, how='left')
    # target encoding (smoothed)
    prior = train_df['Sales'].mean()
    te = (train_df.groupby('Store')['Sales'].agg(['mean','count']))
    smoothing = 10.0
    te['smooth'] = (te['count'] * te['mean'] + smoothing * prior) / (te['count'] + smoothing)
    store_te_map = te['smooth'].to_dict()

    # map back to df (fillna with global prior or feature mean)
    df['store_mean_sales'] = df['Store'].map(sagg['store_mean_sales']).fillna(train_df['Sales'].mean())
    df['store_median_sales'] = df['Store'].map(sagg['store_median_sales']).fillna(train_df['Sales'].median())
    df['store_std_sales'] = df['Store'].map(sagg['store_std_sales']).fillna(0.0)
    if 'store_promo_rate' in sagg.columns or 'Promo' in df.columns:
        df['store_promo_rate'] = df['Store'].map(sagg.get('store_promo_rate', pd.Series())).fillna(0.0)
    df['store_te_sales'] = df['Store'].map(store_te_map).fillna(prior)

    for nm in ['store_mean_sales','store_median_sales','store_std_sales','store_promo_rate','store_te_sales']:
        if nm in missing:
            created.append(nm)

# 5) restore embeddings PCA projection into df if missing
emb_pca_cols = []
if any(c.startswith('store_emb_pca_') for c in missing) or not any(col.startswith('store_emb_pca_') for col in df.columns):
    # try load embeddings artifact
    if os.path.exists(emb_path):
        emb_data = joblib.load(emb_path)
        # emb_data might be dict with 'pca' and 'emb_map' or contain emb_weights
        if isinstance(emb_data, dict) and 'pca' in emb_data and 'emb_map' in emb_data:
            pca = emb_data['pca']
            emb_map = emb_data['emb_map']
            ncomp = pca.n_components_
            emb_pca_cols = [f'store_emb_pca_{i}' for i in range(ncomp)]
            print("Adding emb_pca columns:", emb_pca_cols)
            # map using df.Store -> emb_map (which maps store_id -> vector)
            def get_emb_val(s, idx):
                try:
                    v = emb_map.get(int(s))
                    if v is None:
                        return 0.0
                    return float(v[idx])
                except Exception:
                    return 0.0
            for i,col in enumerate(emb_pca_cols):
                df[col] = df['Store'].map(lambda s: get_emb_val(s,i)).fillna(0.0)
                if col in missing: created.append(col)
        else:
            print("store_embeddings_and_pca.pkl format unexpected; try loading 'emb_weights' and reconstruct PCA.")
            # attempt fallback
            try:
                emb_weights = emb_data['emb_weights']
                from sklearn.decomposition import PCA
                pca = PCA(n_components=min(8, emb_weights.shape[1]), random_state=42)
                emb_proj = pca.fit_transform(emb_weights)
                emb_pca_cols = [f'store_emb_pca_{i}' for i in range(emb_proj.shape[1])]
                # assume store ids map 0..n-1 or shift to 1..n
                min_store = int(df['Store'].min())
                for i,col in enumerate(emb_pca_cols):
                    mapping = { (idx if min_store==0 else idx+1): emb_proj[idx,i] for idx in range(emb_proj.shape[0])}
                    df[col] = df['Store'].map(mapping).fillna(0.0)
                    if col in missing: created.append(col)
            except Exception as e:
                print("Failed to reconstruct embeddings PCA:", e)
    else:
        print("No embeddings artifact found at", emb_path)

# 6) Update FEATURES: keep only columns that now exist in df and remove non-column entries
df_columns = set(df.columns.tolist())
clean_features = [c for c in FEATURES if c in df_columns]
# also append any emb_pca_cols that were created but not in original FEATURES
for c in emb_pca_cols:
    if c not in clean_features and c in df_columns:
        clean_features.append(c)

print("After repair -> present features:", len(clean_features), "| created new features:", len(created))
print("Created features sample:", created[:30])

# 7) Save cleaned FEATURES and updated df back to artifacts (so you don't repeat this)
joblib.dump(clean_features, os.path.join(ARTIFACT_DIR, "features_with_emb.pkl"))
df.to_csv(proc_path, index=False)
print("Saved cleaned FEATURES and updated processed df.")

# 8) Print final sample columns
print("Final FEATURES count:", len(clean_features))
print("Sample final FEATURES:", clean_features[:40])
```

```python
# ===== Final retrain (train+val) -> evaluate on test -> save artifacts =====
import os, joblib, json, time
import pandas as pd, numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
import lightgbm as lgb
import xgboost as xgb

ARTIFACT_DIR = "/content/drive/MyDrive/rossmann_artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

# 1) Load processed df & FEATURES
proc_path = os.path.join(ARTIFACT_DIR, "rossmann_preprocessed.csv")
feat_path = os.path.join(ARTIFACT_DIR, "features_with_emb.pkl")
print("Loading:", proc_path)
df = pd.read_csv(proc_path, parse_dates=['Date'], low_memory=False)
FEATURES = joblib.load(feat_path)
print("Loaded df shape:", df.shape, "| FEATURES:", len(FEATURES))

# 2) Recreate chronological splits (train+val as full train, test as after DATE_SPLIT)
DATE_SPLIT = "2015-06-01"  # use your DATE_SPLIT (adjust if you want)
train_df = df[df['Date'] < DATE_SPLIT].sort_values('Date').reset_index(drop=True)
test_df  = df[df['Date'] >= DATE_SPLIT].sort_values('Date').reset_index(drop=True)

# Use entire train_df (train+val) for final training
X_train_full = train_df[FEATURES]
y_train_full = np.log1p(train_df['Sales'].values)
X_test = test_df[FEATURES]
y_test_raw = test_df['Sales'].values
y_test = np.log1p(y_test_raw)

print("Train full shape:", X_train_full.shape, "Test shape:", X_test.shape)

# 3) LightGBM final train (use previous tuned params if present)
lgb_params = {
    'objective':'regression','metric':'rmse','learning_rate':0.05,
    'num_leaves':127,'min_data_in_leaf':50,'verbosity':-1,'n_jobs':-1,'seed':42
}
# if you have tuned params saved earlier (optuna), you can load and override
opt_lgb_path = os.path.join(ARTIFACT_DIR, "optuna_lgb_study_final.pkl")
try:
    import joblib as _jl
    if os.path.exists(opt_lgb_path):
        study = _jl.load(opt_lgb_path)
        best = study.best_trial.params
        # map typical params (example)
        lgb_params.update({
            'learning_rate': best.get('learning_rate', lgb_params['learning_rate']),
            'num_leaves': int(best.get('num_leaves', lgb_params['num_leaves'])),
            'min_data_in_leaf': int(best.get('min_data_in_leaf', lgb_params['min_data_in_leaf'])),
            'feature_fraction': best.get('feature_fraction', 1.0),
            'bagging_fraction': best.get('bagging_fraction', 1.0),
            'bagging_freq': int(best.get('bagging_freq', 0))
        })
        print("Loaded tuned LGB params from Optuna study.")
except Exception:
    pass

print("Training final LightGBM...")
t0 = time.time()
lgb_train_full = lgb.Dataset(X_train_full, label=y_train_full)
gbm_final = lgb.train(lgb_params, lgb_train_full, num_boost_round=1500, valid_sets=[lgb_train_full],
                      callbacks=[lgb.log_evaluation(0)])
t1 = time.time()
print(f"LightGBM trained in {int(t1-t0)}s | best_iter: {getattr(gbm_final,'best_iteration',None)}")

# 4) XGBoost final train
xgb_params = {
    'objective':'reg:squarederror','eta':0.05,'max_depth':8,'subsample':0.8,'colsample_bytree':0.8,'verbosity':0,'seed':42
}
print("Training final XGBoost...")
t0 = time.time()
dtrain = xgb.DMatrix(X_train_full, label=y_train_full)
bst_final = xgb.train(xgb_params, dtrain, num_boost_round=1000, evals=[(dtrain,'train')],
                      verbose_eval=False)
t1 = time.time()
print(f"XGBoost trained in {int(t1-t0)}s | best_ntree_limit: {getattr(bst_final,'best_iteration',None) or getattr(bst_final,'best_ntree_limit',None)}")

# 5) Predict on test and evaluate (log-scale -> real-scale)
pred_lgb_test_log = gbm_final.predict(X_test, num_iteration=getattr(gbm_final,'best_iteration',None))
pred_xgb_test_log = bst_final.predict(xgb.DMatrix(X_test), iteration_range=(0, getattr(bst_final,'best_iteration',None) or None))

pred_lgb_test = np.expm1(pred_lgb_test_log)
pred_xgb_test = np.expm1(pred_xgb_test_log)

# simple inverse-RMSE weighting ensemble
rmse_lgb_val = np.sqrt(mean_squared_error(np.expm1(y_train_full[:1000]), np.expm1(gbm_final.predict(X_train_full[:1000]))))  # quick proxy on subset
rmse_xgb_val = np.sqrt(mean_squared_error(np.expm1(y_train_full[:1000]), np.expm1(bst_final.predict(xgb.DMatrix(X_train_full[:1000])))))
w_l = 1.0/(rmse_lgb_val + 1e-9); w_x = 1.0/(rmse_xgb_val + 1e-9)
pred_ens_test = (w_l*pred_lgb_test + w_x*pred_xgb_test)/(w_l+w_x)

rmse_ens = np.sqrt(mean_squared_error(y_test_raw, pred_ens_test))
mae_ens = mean_absolute_error(y_test_raw, pred_ens_test)
print(f"FINAL Ensemble Test RMSE: {rmse_ens:.2f} | MAE: {mae_ens:.2f}")

# 6) Save models and metrics
joblib.dump(gbm_final, os.path.join(ARTIFACT_DIR, "lgb_final_with_emb.pkl"))
bst_final.save_model(os.path.join(ARTIFACT_DIR, "xgb_final_with_emb.json"))
joblib.dump(FEATURES, os.path.join(ARTIFACT_DIR, "features_with_emb.pkl"))

metrics = {
    "date": str(pd.Timestamp.now()),
    "date_split": DATE_SPLIT,
    "rmse_ens_test": float(rmse_ens),
    "mae_ens_test": float(mae_ens)
}
with open(os.path.join(ARTIFACT_DIR, "final_models_metrics.json"), "w") as f:
    json.dump(metrics, f, indent=2)

# 7) Save test predictions (optional)
test_out = test_df.copy()
test_out['pred_ens'] = pred_ens_test
test_out.to_csv(os.path.join(ARTIFACT_DIR, "test_predictions_final.csv"), index=False)

print("Saved final models, metrics and test predictions to:", ARTIFACT_DIR)
```
